```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('2019.csv', delimiter=',')

df19 = df.copy()
df19 = df19.set_index('Overall rank')
df19.head()
```

#### 1. Untuk setiap numerik feature/column (kecuali column Overall Rank), hitunglah:

```python
# MEAN
mean = df19.mean(numeric_only=True)
mean
```

```python
# MEDIAN
median = df19.median(numeric_only=True)
median
```

```python
# MODUS
mode = df19.mode(numeric_only=True)
mode
```

```python
# Range = max - min
r = df19.max(numeric_only=True) - df19.min(numeric_only=True)
r
```

```python
# Q1
q1 = df19.quantile(0.25, numeric_only=True)
q1
```

```python
# Q3
q3 = df19.quantile(0.75, numeric_only=True)
q3
```

```python
# IQR = Q3 - Q1
iqr = q3 - q1
iqr
```

```python
# Std
std = df19.std(ddof=1)
std
```

```python
# Variance = std^2
var = df19.var()
var
```

```python
# Z-score
df19['Z-Score'] = (df19['Score'] - df19['Score'].mean()) / df19['Score'].std(ddof=1)
df19['Z-GDP'] = (df19['GDP per capita'] - df19['GDP per capita'].mean()) / df19['GDP per capita'].std(ddof=1)
df19['Z-Social'] = (df19['Social support'] - df19['Social support'].mean()) / df19['Social support'].std(ddof=1)
df19['Z-Health'] = (df19['Healthy life expectancy'] - df19['Healthy life expectancy'].mean()) / df19['Healthy life expectancy'].std(ddof=1)
df19['Z-Freedom'] = (df19['Freedom to make life choices'] - df19['Freedom to make life choices'].mean()) / df19['Freedom to make life choices'].std(ddof=1)
df19['Z-Generosity'] = (df19['Generosity'] - df19['Generosity'].mean()) / df19['Generosity'].std(ddof=1)
df19['Z-Corruption'] = (df19['Perceptions of corruption'] - df19['Perceptions of corruption'].mean()) / df19['Perceptions of corruption'].std(ddof=1)
df19
```

#### 2. Untuk setiap numerik feature/column (kecuali column Overall Rank), carilah data outlier-nya berdasarkan:

```python
# IQR Method
high_out = q3 + 1.5 * iqr 
low_out = q1 - 1.5 * iqr 

(df19 < low_out)|(df19 > high_out)
# True = outliers
```

```python
# Z-Score Method
highout = mean + 2 * std
lowout = mean - 2 * std

(df19 < lowout)|(df19 > highout)
# True = outliers
```

#### 3. Tampilkan sebaran masing-masing numerik feature/column (kecuali column Overall Rank) dalam bentuk boxplot dan histogram!

```python
# Box Plot
plt.subplots(figsize=(15, 10))
plt.boxplot([df['Score'], df19['GDP per capita'], df19['Social support'], df['Healthy life expectancy'],
              df['Freedom to make life choices'], df['Generosity'], df['Perceptions of corruption']], vert=True)
plt.xticks([1,2,3,4,5,6,7],
    ['Score', 'GDP per Capita', 'Social Support', 'Healthy Life Expectancy', 'Freedom to Make \n Life Choices', 'Generosity', 'Perceptions of Corruption']
)
plt.title('Box Plot - Sebaran Numerik Feature/Column', fontsize=20, pad=10)
plt.show()
```

```python
# Histogram
minval = df19.min(numeric_only=True)
maxval = df19.max(numeric_only=True)
```

```python
binsScore = np.linspace(minval['Score'] - 0.005, maxval['Score'] + 0.005, 10)
binsGDP = np.linspace(minval['GDP per capita'] - 0.005, maxval['GDP per capita'] + 0.005, 10)
binsSoc = np.linspace(minval['Social support'] - 0.005, maxval['Social support'] + 0.005, 10)
binsHealth = np.linspace(minval['Healthy life expectancy'] - 0.005, maxval['Healthy life expectancy'] + 0.005, 10)
binsFree = np.linspace(minval['Freedom to make life choices'] - 0.005, maxval['Freedom to make life choices'] + 0.005, 10)
binsGen = np.linspace(minval['Generosity'] - 0.005, maxval['Generosity'] + 0.005, 10)
binsCorr = np.linspace(minval['Perceptions of corruption'] - 0.005, maxval['Perceptions of corruption'] + 0.005, 10)

plt.figure(figsize=(20,15))

plt.subplot(331)
plt.hist(df['Score'], binsScore)
plt.title('Score Histogram')
plt.xticks(binsScore, rotation=45)

plt.subplot(332)
plt.hist(df['GDP per capita'], binsGDP)
plt.xticks(binsGDP, rotation=45)
plt.title('GDP per Capita Histogram')

plt.subplot(333)
plt.hist(df['Social support'], binsSoc)
plt.xticks(binsSoc, rotation=45)
plt.title('Social Support')

plt.subplot(334)
plt.hist(df['Healthy life expectancy'], binsSoc)
plt.xticks(binsHealth, rotation=45)
plt.title('Healthy Life Expectancy')

plt.subplot(335)
plt.hist(df['Freedom to make life choices'], binsFree)
plt.title('Freedom to Make Life Choices Histogram')
plt.xticks(binsFree, rotation=45)

plt.subplot(336)
plt.hist(df['Generosity'], binsGen)
plt.title('Generosity Histogram')
plt.xticks(binsGen, rotation=45)

plt.subplot(337)
plt.hist(df['Perceptions of corruption'], binsCorr)
plt.title('Perceptions of Corruption Histogram')
plt.xticks(binsCorr, rotation=45)

plt.suptitle('Histogram - Sebaran Numerik Feature/Column', fontsize=20)

plt.show()
```

#### 4. Di antara semua numerik feature/column (kecuali column Overall Rank), feature/column mana saja yang sangat mempengaruhi happiness score suatu negara? Buktikan dengan menghitung nilai:

<li>Covariance
<li>Pearson Correlation
<li>Spearman Correlation
<li>Kendall Correlation

```python
# Answer = GDP per capita, Social Support, dan Healthy life expectancy
# karena nilai correlationnya kuat (mendekati 1) dan nilai covariancenya positif maka semakin besar nilai ketiga kolom tsb, nilai kolom Score akan semakin meningkat
```

```python
cov = df19.cov()
cov
```

```python
pearson_corr =  df19.corr(method='pearson')
pearson_corr
```

```python
spearman_corr =  df19.corr(method='spearman')
spearman_corr
```

```python
kendall_corr =  df19.corr(method='kendall')
kendall_corr
```

#### 5. Visualisasikan dalam bentuk bar chart data berikut:

<li>10 Negara dengan GDP per capita tertinggi.
<li>10 Negara dengan Healthy life expectancy tertinggi.
<li>10 Negara dengan Perceptions of corruption tertinggi.

```python
dfgdp = df19.copy()
dfgdp = dfgdp.sort_values('GDP per capita', ascending=False).head(10)

dfhealth = df19.copy()
dfhealth = dfhealth.sort_values('Healthy life expectancy', ascending=False).head(10)

dfcorr = df19.copy()
dfcorr = dfcorr.sort_values('Perceptions of corruption', ascending=False).head(10)
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(20,5))

plt.subplot(131)
barGDP = plt.bar(dfgdp['Country or region'], dfgdp['GDP per capita'], width=.5,
                color=['tab:blue'])
plt.title('10 Negara dengan GDP per Capita Tertinggi', fontsize=15)
plt.ylabel('GDP per Capita', fontsize=12)
plt.xlabel('Negara', fontsize=12)
plt.xticks(rotation=90)

plt.subplot(132)
barHealth = plt.bar(dfhealth['Country or region'], dfhealth['Healthy life expectancy'], width=.5,
                    color=['tab:green'])
plt.title('10 Negara dengan Healthy Life Expectancy Tertinggi',fontsize=15)
plt.ylabel('Healthy Life Expectancy', fontsize=12)
plt.xlabel('Negara', fontsize=12)
plt.xticks(rotation=90)

plt.subplot(133)
barCorrupt = plt.bar(dfcorr['Country or region'], dfcorr['Perceptions of corruption'], width=.5,
                color=['tab:olive'])
plt.title('10 Negara dengan Perceptions of Corruption Tertinggi', fontsize=15)
plt.ylabel('Perceptions of Corruption', fontsize=12)
plt.xlabel('Negara', fontsize=12)
plt.xticks(rotation=90)

plt.show()
```
